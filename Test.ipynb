```python
# https://github.com/opencv/opencv/tree/master/data/haarcascades
```

```python
faceCascade_default = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
#faceCascade_alt = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_alt.xml')
```

```python
my_face = faceCascade_default.detectMultiScale(img,)
for x,y,w,h in my_face:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
cv2_imshow(img)
```
